```python
import numpy
import matplotlib
%matplotlib inline
import glob
```

```python
# a function which takes in a .csv file and plots the
# mean, max and min
def analyze(filename):

    # Load file
    data = numpy.loadtxt(fname=filename, delimiter=",")

    # Create matplotlib figure
    figure = matplotlib.pyplot.figure(figsize=(10, 3))

    # Add plots to figure
    subplot1 = figure.add_subplot(1, 3, 1)
    subplot2 = figure.add_subplot(1, 3, 2)
    subplot3 = figure.add_subplot(1, 3, 3)

    # Subplot 1
    subplot1.set_ylabel("average")
    subplot1.plot(data.mean(axis=0))

    # Subplot 2
    subplot2.set_ylabel("max")
    subplot2.plot(data.max(axis=0))

    # Subplot 3
    subplot3.set_ylabel("min")
    subplot3.plot(data.min(axis=0))

    # Configure layout
    figure.tight_layout()

    # Display figure
    matplotlib.pyplot.show(figure)
```

```python
def detect_problems(filename):
    #load data
    data = numpy.loadtxt(fname=filename, delimiter=',')
    
    if data.max(axis=0)[0] == 0 and data.max(axis=0)[20]== 20:
        print("Suspicious looking maxima!")
    elif data.min(axis=0).sum() == 0:
        print("Minima add up to zero!")
    else:
        print("Seems OK!")
        
```

```python
# makes a list of the filenames
filenames = glob.glob('inflammation*.csv')

# loop over all the filenames and print file, plot data 
# and flag suspicious data
for files in filenames:
    # printing the files to act as a label
    print(files)
    # a function which takes in a .csv file and plots the
    # mean, max and min
    analyze(files)
    # a function checks for suspicious maxima and minima in 
    # the dataset
    detect_problems(files)
```
